```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a DataFrame with the provided data
data = {
    'Model': [
        'KNN', 'Decision Tree', 'Logistic Regression', 'Naive Bayes',
        'SVM', 'Random Forest', 'Ada Boost', 'Gradient Boosting',
        'DNN', 'Transformer'
    ],
    'Original': [
        0.84, 0.82, 0.84, 0.86, 0.86, 0.89, 0.86, 0.91,
        0.84, 0.89
    ],
    'Improved': [
        0.81, 0.91, 0.91, 0.93, 0.91, 0.95, 0.93, 0.95,
        0.93, 0.95
    ]
}

df = pd.DataFrame(data)
df = df.melt('Model', var_name='Version', value_name='Accuracy')
# Fixing text positioning for values on the bars
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Drawing a barplot
ax = sns.barplot(x='Accuracy', y='Model', hue='Version', data=df, palette="muted")

# Correcting the text placement at the end of the bars
for p in ax.patches:
    width = p.get_width()
    if width > 0:  # Ensure only positive values are shown
        plt.text(
            width,  # Exact position at the bar's end
            p.get_y() + p.get_height() / 2,  # Vertically centered
            f'{width:.2f}',  # Format value to two decimal places
            ha='left', va='center', fontsize=10  # Alignment and font size
        )

# Setting the labels and title
plt.title('Model Accuracy Comparison (Original vs Improved)', fontsize=20)
plt.xlabel('Accuracy', fontsize=15)
plt.ylabel('Model', fontsize=15)

# Positioning the legend
plt.legend(title='Version', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Show the plot
plt.tight_layout()
plt.show()

```
